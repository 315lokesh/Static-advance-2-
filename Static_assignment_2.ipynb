Que.4 How do you interpret the P-value in hypothesis testing?

Ans: represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true

1. Understanding the Basics:
Null Hypothesis:

This is a statement of no effect or no difference that we are trying to disprove.

Alternative Hypothesis:
This is the statement we are trying to find evidence for, suggesting an effect or difference.

P-value:
It's the probability of observing data as extreme as, or more extreme than, what was observed in the sample data, assuming the null hypothesis is true.

2. Interpreting the P-value:
Small p-value (typically ≤ 0.05):
Indicates strong evidence against the null hypothesis. It suggests that the observed results are unlikely to have occurred by chance if the null hypothesis were true, and therefore, we reject the null hypothesis.
Large p-value (typically > 0.05):
Indicates weak evidence against the null hypothesis. It suggests that the observed results are likely to have occurred by chance if the null hypothesis were true, and therefore, we fail to reject the null hypothesis.

3. Significance Level (alpha):
The significance level (alpha) is a threshold (usually 0.05) that we set before conducting the hypothesis test.
If the p-value is less than or equal to alpha, we reject the null hypothesis. If it's greater than alpha, we fail to reject the null hypothesis.

4. Important Considerations:
P-value doesn't prove the alternative hypothesis:
Rejecting the null hypothesis doesn't definitively prove the alternative hypothesis is true, it just means there's enough evidence to suggest it.
Context matters:
The choice of significance level (alpha) can vary depending on the field of study and the consequences of making a wrong decision.
P-value is not the probability of the null hypothesis being true:
Wikipedia clarifies that the p-value is not the probability that the null hypothesis is true. It's the probability of the observed data (or more extreme data) given the null hypothesis.

Que.5 What are Type 1 and Type 2 errors in hypothesis testing?

Ans:a Type I error occurs when a true null hypothesis is incorrectly rejected, leading to a false positive. A Type II error happens when a false null hypothesis is not rejected, resulting in a false negative.

1. Null Hypothesis:
In hypothesis testing, we start with a null hypothesis (H0), which is a statement about the population that we assume to be true until evidence suggests otherwise.

2. Type I Error (False Positive):
Definition: Rejecting the null hypothesis when it is actually true.
Example: If a medical test incorrectly indicates a person has a disease (when they don't), that's a Type I error.
Probability: The probability of making a Type I error is denoted by alpha (α), often set at 0.05

3. Type II Error (False Negative):
Definition: Failing to reject the null hypothesis when it is actually false.
Example: If a medical test incorrectly indicates a person does not have a disease (when they do), that's a Type II error.
Probability: The probability of making a Type II error is denoted by beta (β).

4. Relationship between Type I and Type II Errors:
There's an inverse relationship between the two. Reducing the chance of one type of error often increases the chance of the other.
For example, lowering the significance level (alpha) to reduce Type I errors can increase the risk of Type II errors.

5. Importance of Power:
The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.
High power is desirable because it means the test is more likely to detect a real effect.

Que.8 What is the Z-test, and when is it used in hypothesis testing8

Assumes a normal distribution:
Z-tests are most reliable when the data being analyzed is normally distributed or when the sample size is large enough for the central limit theorem to apply.

Requires knowledge of population variance (or large sample size):
Z-tests require either the population standard deviation (or variance) to be known, or a large sample size (typically n ≥ 30) where the sample standard deviation can be used as an estimate of the population standard deviation.

Compares means:

Z-tests are used to compare a sample mean to a known population mean (one-sample z-test) or to compare the means of two independent samples.

Calculates a z-statistic:
The z-test calculates a z-statistic, which represents the number of standard deviations the sample mean is away from the population mean.

Compares z-statistic to a critical value:
The calculated z-statistic is then compared to a critical value from the standard normal distribution to determine the statistical significance of the difference.

When to use a z-test:

Large sample sizes:
When working with a large sample (n ≥ 30), a z-test can be used even if the population variance is unknown, as the sample variance can be used as an estimate.

Known population variance:
If the population standard deviation (or variance) is known, a z-test can be used for both large and small samples.

Comparing sample mean to population mean:
When you want to compare a sample's average to a known population average.

Comparing means of two independent samples:
When you want to compare the averages of two independent groups and you know
the population variance or have large samples.

In essence, the z-test is a powerful tool for hypothesis testing, allowing researchers to make inferences about populations based on sample data when certain conditions are met.

Que.9 How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans: A Z-score measures how many standard deviations a data point is from the mean of a dataset. In hypothesis testing, it's used to determine the probability of obtaining a result as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. It helps determine if a sample mean is significantly different from a population mean.

**Calculating the Z-score:**

The Z-score is calculated using the following formula:
Z = (x - μ) / σ

**Where:**

x: is the data point (or sample mean)
μ: is the population mean
σ: is the population standard deviation
If you are working with a sample and don't know the population standard deviation, you can use the sample standard deviation (s) instead:
Z = (x - x̄) / s

Where:
**bold text**x̄: is the sample mean

s: is the sample standard deviation
Z-score in Hypothesis Testing:

In hypothesis testing, a Z-test is used when you know the population standard deviation or have a large sample size. The Z-score helps determine if the observed difference between a sample mean and a hypothesized population mean is statistically significant.


Here's how it works:

1. State the null and alternative hypotheses:
The null hypothesis (H0) assumes no significant difference, while the
alternative hypothesis (Ha) suggests there is a difference.

2. Calculate the Z-score:
Using the formula above, calculate the Z-score for your sample data.

3. Determine the critical value:
Based on your chosen significance level (alpha, usually 0.05) and whether it's a one-tailed or two-tailed test, find the critical Z-value from a Z-table.

4. Compare the Z-score to the critical value:
If the absolute value of the calculated Z-score is greater than the critical Z-value, you reject the null hypothesis. This indicates that the observed difference is statistically significant. If the absolute value of the Z-score is less than the critical value, you fail to reject the null hypothesis.

The shape of the t-distribution is determined by its degrees of freedom, which are related to the sample size.
As the sample size (and thus the degrees of freedom) increases, the t-distribution gradually approaches the normal distribution.

**When to use the t-distribution instead of the normal distribution: **
**Small sample sizes: **

Unknown population standard deviation:
If the population standard deviation is unknown, the t-distribution is used to account for the uncertainty in estimating it from the sample standard deviation.

Confidence intervals and hypothesis testing:
The t-distribution is used in constructing confidence intervals for population means and in performing t-tests, which are used to compare the means of two samples.

Que.18 What is the Chi-square goodness of fit test, and how is it applied?

Ans:- 1. Define the Hypotheses:
Null Hypothesis (H0): The observed data follows the expected distribution.
Alternative Hypothesis (Ha): The observed data does not follow the expected distribution.

2. Determine Expected Frequencies:
For each category in your observed data, calculate the expected frequency based on the hypothesized distribution.
If no specific distribution is hypothesized, equal probabilities (e.g., a uniform distribution) are often assumed.

3. Calculate the Chi-Square Statistic:
The chi-square statistic is calculated using the following formula: χ² = Σ [(Observed - Expected)² / Expected].
Sum the results for each category.

4. Determine Degrees of Freedom:
Degrees of freedom (df) are calculated as (number of categories - 1).

5. Determine the p-value:
Using the calculated chi-square statistic and degrees of freedom, find the p-value from a chi-square distribution table or using statistical software.
The p-value represents the probability of observing a chi-square statistic as extreme as, or more extreme than, the one calculated, if the null hypothesis is true.

6. Make a Decision:
Compare the p-value to a pre-determined significance level (alpha, typically 0.05).

If the p-value is less than or equal to alpha, reject the null hypothesis and conclude that the observed distribution is significantly different from the expected distribution.
If the p-value is greater than alpha, fail to reject the null hypothesis and conclude that there is not enough evidence to say the observed distribution is significantly different from the expected distribution.


Que.19 What is an ANOVA test, and what are its assumptions?

Ans: An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there is a statistically significant difference between them. It's essentially an extension of the t-test for comparing more than two means. The core idea is to partition the total variability in the data into different sources of variation, specifically between groups and within groups.

Assumptions of ANOVA:

ANOVA relies on several key assumptions for its results to be valid:

1. Independence of Observations:
The data points within each group, as well as between groups, should be independent of each other. This means that the value of one observation should not influence the value of another.

2. Normality:
The data within each group should be approximately normally distributed. This assumption is more crucial for smaller sample sizes.

3. Homogeneity of Variance (also known as Homoscedasticity):
The variances of the data within each group should be approximately equal. In other words, the spread of the data should be similar across all groups being compared.



Que.21 What are the different types of ANOVA tests?

Ans: Different types of ANOVA (Analysis of Variance) tests are used to compare the means of two or more groups to determine if there's a statistically significant difference between them. The choice of ANOVA test depends on the number of independent variables (factors) and dependent variables in the research design.

Here's an overview of the main types of ANOVA tests:

1. One-way ANOVA
Purpose: Compares the means of three or more independent groups based on a single independent variable (factor).
Example: Comparing the mean test scores of students from three different schools, based on a single factor like teaching method (e.g., traditional, online, hybrid).

Assumptions: The data within each group should be approximately normally distributed, the variances of the groups should be roughly equal (homogeneity of variance), and the observations within each group should be independent.
Note: If there are only two groups to compare, a t-test is typically used instead of a one-way ANOVA.

2. Two-way ANOVA
Purpose: Examines the effect of two independent variables and their potential interaction on a single dependent variable.
Example: Analyzing student performance based on teaching method and gender.
Assumptions: Includes independence of observations, normality, homogeneity of variances, and consistency of interaction effects across levels.

3. Factorial ANOVA
Purpose: An extension of two-way ANOVA that handles multiple independent variables.
Example: Studying the impact of teaching method, grade level, and type of music on student test scores.
Note: Analyzing many factors can be complex.

4. Repeated measures ANOVA
Purpose: Used for data collected from the same participants under different conditions or over time.
Example: Measuring cholesterol levels of the same patients at different time points after a dietary change.
Assumptions: Includes sphericity in addition to normality, homogeneity of variance, and independence.

5. Multivariate analysis of variance (MANOVA)
Purpose: Used to analyze the effects of one or more independent variables on multiple dependent variables simultaneously.
Example: Investigating how fertilizer impacts both crop yield and plant height.
Assumptions: Includes multivariate normality and homogeneity of variance-covariance matrices.

6. Mixed ANOVA
Purpose: Used when a study includes both independent groups and repeated measurements on those groups.
Example: Comparing two treatments on two groups, measuring symptoms over time within each group.


Que.22 What is the F-test, and how does it relate to hypothesis testing?

Ans: The F-test is a statistical test commonly used in hypothesis testing to compare the variances of two populations or assess the significance of a regression model.

the F-test relates to hypothesis testing

Formulating Hypotheses: Like other hypothesis tests, the F-test involves stating a null hypothesis (H₀) and an alternative hypothesis (H₁). The H₀ for an F-test typically states that the variances of the two populations are equal, or that a regression model has no significant relationship between variables. The H₁ proposes that there's a significant difference in variances or a significant relationship in the model.

Calculating the F-statistic: The test calculates an F-statistic, which is the ratio of the two variances being compared.

Determining the Critical Value: The calculated F-statistic is then compared to a critical value obtained from an F-distribution table. This table requires the degrees of freedom for the numerator and denominator, as well as the chosen significance level (alpha, α), which represents the probability of rejecting the null hypothesis when it is actually true.

Decision-Making: The decision to reject or fail to reject the null hypothesis is based on comparing the calculated F-statistic to the critical value. If the F-statistic exceeds the critical value, the null hypothesis is rejected, indicating a statistically significant difference in variances or a significant relationship in the regression model.

```python



def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x : int : The outcome (0 or 1)
    p : float : Probability of success (0 <= p <= 1)

    Returns:
    float : Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1")

    return p**x * (1 - p)**(1 - x)

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.7))  # Success
print("P(X=0):", bernoulli_pdf(0, 0.7))  # Failure
```

```python
#9
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a dataset and plots them.

    Parameters:
    data (list or array): The dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='blue', label='Z-scores')
    plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
    plt.axhline(y=1, color='green', linestyle='--', label='±1σ')
    plt.axhline(y=-1, color='green', linestyle='--')
    plt.axhline(y=2, color='orange', linestyle='--', label='±2σ')
    plt.axhline(y=-2, color='orange', linestyle='--')
    plt.axhline(y=3, color='red', linestyle='--', label='±3σ')
    plt.axhline(y=-3, color='red', linestyle='--')

    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(50, 10, 100)  # Sample data from normal distribution
z = calculate_and_plot_z_scores(data)
```

```python
#4
import numpy as np
from scipy.stats import norm

# Sample data (e.g., test scores)
sample = [68, 72, 75, 70, 69, 74, 71, 73, 68, 70]

# Known population parameters
mu_0 = 70       # Hypothesized population mean
sigma = 3       # Known population standard deviation
n = len(sample)

# Step 1: Calculate sample mean
x_bar = np.mean(sample)

# Step 2: Calculate Z-statistic
z = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Step 3: Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print results
print(f"Sample Mean = {x_bar:.2f}")
print(f"Z-statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")

# Step 4: Conclusion at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0)")
else:
    print("Fail to reject the null hypothesis (H0)")
```

```python
#5
import numpy as np
from scipy import stats

# Sample dataset
data = [54, 49, 51, 52, 50, 48, 53, 47, 49, 51]

# Step 1: Sample statistics
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)  # Sample standard deviation

# Step 2: Confidence level
confidence = 0.95
alpha = 1 - confidence
df = n - 1  # degrees of freedom
t_critical = stats.t.ppf(1 - alpha/2, df)

# Step 3: Margin of error
margin_of_error = t_critical * (std / np.sqrt(n))

# Step 4: Confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Output
print(f"Sample Mean = {mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#13
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_skew_kurt(data=None, size=1000, mean=0, std=1):
    """
    Calculates skewness and kurtosis of a dataset (normal by default)
    """
    if data is None:
        data = np.random.normal(loc=mean, scale=std, size=size)

    skew_val = skew(data)
    kurt_val = kurtosis(data)  # Fisher's definition: normal dist → 0

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis: {kurt_val:.4f} (0 means normal distribution)")

    return skew_val, kurt_val

# Example use:
analyze_distribution_skew_kurt()
```

```python
#16
 import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Print results
    print(f"Sample Mean       = {sample_mean:.2f}")
    print(f"Population Mean   = {population_mean}")
    print(f"Z-statistic       = {z:.4f}")
    print(f"P-value           = {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print(" Reject the null hypothesis (Significant difference)")
    else:
        print(" Fail to reject the null hypothesis (No significant difference)")
# Example
np.random.seed(0)
sample = np.random.normal(loc=105, scale=10, size=30)  # Simulated sample data
population_mean = 100
population_std = 10

one_sample_z_test(sample, population_mean, population_std)
```

```python
#19
import numpy as np
from scipy.stats import norm

# Step 1: Given Data
population_mean = 1000       # μ
sample_mean = 980            # x̄
std_dev = 100                # σ (population standard deviation)
sample_size = 50             # n
alpha = 0.05                 # Significance level (5%)

# Step 2: Calculate Z-statistic
z = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
print(f"Z-statistic: {z:.2f}")

# Step 3: Critical Z-value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)
print(f"Critical Z-value (±): {z_critical:.2f}")

# Step 4: Conclusion
if abs(z) > z_critical:
    print(" Reject the Null Hypothesis (Significant difference)")
else:
    print(" Fail to Reject the Null Hypothesis (No significant difference)")
```

```python
#20
import numpy as np
import scipy.stats as stats

# Step 1: Sample dataset
data = [65, 70, 72, 68, 74, 71, 69, 73, 75, 70]

# Step 2: Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

# Step 3: Confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 4: t-critical value (because population std is unknown)
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 5: Margin of Error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 6: Confidence Interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 7: Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"We are 95% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}")
```

```python
#21
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 170          # population mean (e.g., height in cm)
sigma = 10        # population std deviation
n = 50            # sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 3: t-critical value (since sigma unknown)
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Step 4: Margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 5: Confidence Interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 6: Output results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 7: Optional - Visualize
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='Lower CI')
plt.axvline(ci_upper, color='green', linestyle='--', label='Upper CI')
plt.title("Sample Data Histogram with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()
```

```python
#23
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Define parameters of the normal distribution
mu = 0       # Mean
sigma = 1    # Standard deviation

# Step 2: Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Step 3: Calculate the PDF values
pdf = norm.pdf(x, mu, sigma)

# Step 4: Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, color='blue', label=f'N({mu}, {sigma}²)')
plt.title('Normal Distribution - PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.fill_between(x, pdf, alpha=0.3, color='skyblue')
plt.show()
```

```python
#23
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Define Poisson distribution parameter
lam = 4  # λ = average rate (e.g., 4 events per hour)

# Step 2: Define range of x values (number of events)
x = np.arange(0, 15)

# Step 3: Calculate CDF values
cdf = poisson.cdf(x, mu=lam)

# Step 4: Display values
for i in range(len(x)):
    print(f"P(X ≤ {x[i]}) = {cdf[i]:.4f}")

# Step 5: Plot the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.title(f'Poisson CDF (λ = {lam})')
plt.xlabel('Number of Events (x)')
plt.ylabel('Cumulative Probability')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

```python
#24
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Set parameters for uniform distribution
a = 10   # lower bound
b = 20   # upper bound
n = 10000  # number of simulations

# Step 2: Simulate random variable
data = np.random.uniform(low=a, high=b, size=n)

# Step 3: Calculate expected value (mean)
expected_value = np.mean(data)
print(f"Expected Value (Simulated Mean): {expected_value:.2f}")

# Theoretical expected value of Uniform(a, b) = (a + b) / 2
theoretical_mean = (a + b) / 2
print(f"Theoretical Mean: {theoretical_mean:.2f}")

# Step 4: Plot the distribution
plt.hist(data, bins=30, color='skyblue', edgecolor='black', density=True)
plt.axvline(expected_value, color='red', linestyle='--', label='Simulated Mean')
plt.axvline(theoretical_mean, color='green', linestyle='--', label='Theoretical Mean')
plt.title('Continuous Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
#26
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # Reproducibility
mu = 100           # Mean
sigma = 15         # Standard deviation
size = 1000        # Number of data points

data = np.random.normal(loc=mu, scale=sigma, size=size)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Step 4: Output results
print(f"Range: {data_range:.2f}")
print(f"Q1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
print(f"IQR (Q3 - Q1): {iqr:.2f}")

# Step 5: Optional - Boxplot to visualize IQR
plt.boxplot(data, vert=False)
plt.title('Boxplot of Normally Distributed Data')
plt.xlabel('Value')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#27
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a normal dataset (e.g., test scores)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=1000)

# Step 2: Calculate mean and std deviation
mean = np.mean(data)
std = np.std(data)

# Step 3: Apply Z-score normalization
z_scores = (data - mean) / std

# Step 4: Check stats of normalized data
print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
print(f"Z-score Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")

# Step 5: Visualize before and after normalization
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.4)

# Z-score normalized histogram
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
#28
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution(mu=0, sigma=1, size=1000):
    """
    Generates a normal distribution and calculates skewness & kurtosis.
    """
    # Step 1: Generate normal data
    data = np.random.normal(loc=mu, scale=sigma, size=size)

    # Step 2: Calculate skewness and kurtosis
    skewness = skew(data)
    kurt = kurtosis(data, fisher=False)  # Fisher=False → Pearson's definition (normal = 3)

    # Step 3: Print results
    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis: {kurt:.4f}")

    # Step 4: Interpretation
    if abs(skewness) < 0.5:
        print(" The data is approximately symmetric.")
    elif skewness > 0:
        print(" The data is positively skewed (right tail).")
    else:
        print(" The data is negatively skewed (left tail).")

    if kurt > 3:
        print(" The data is leptokurtic (sharper peak).")
    elif kurt < 3:
        print(" The data is platykurtic (flatter peak).")
    else:
        print(" The data has normal kurtosis (mesokurtic).")

# Call the function
analyze_distribution()
```
